## Libraries

```python
import numpy as np
import pandas as pd
import scipy

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.preprocessing import StandardScaler

from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import KMeans

from sklearn.decomposition import PCA
import pickle
```

## Standardization

```python
scaler = StandardScaler()
segmentation_std = scaler.fit_transform(df_segmentation)
```

## Hierachical Clustering
* Determine the number of clusters
* Use flat clustering for segmentation

```python
hier_clust = linkage(segmentation_std, method= 'ward')
```

```python
plt.figure(figsize = (12,9))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Observations')
plt.ylabel('Distance')
dendrogram(hier_clust,
           truncate_mode = 'level',
           p = 5,
          show_leaf_counts = False,
          no_labels = True,
          )
plt.show()
```

## K-means Clustering
* For data segmentation

```python
#Within cluster sum of squares values to determine the best clustering solution.
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init='k-means++', random_state= 42)
    kmeans.fit(segmentation_std)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize = (10,8))
plt.plot(range(1,11), wcss, marker = 'o', linestyle = '--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('K-means Clustering')
plt.show()
```

### Number of clusters obtained is 4 using the elbow method

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
```

```python
kmeans.fit(segmentation_std)
```

## Results

```python
df_segm_kmeans = df_segmentation.copy()
df_segm_kmeans['Segment K-means'] = kmeans.labels_
```

```python
df_segm_analysis = df_segm_kmeans.groupby(['Segment K-means']).mean()
df_segm_analysis
```

### For segment 0: 0.5 for Sex indicates 50% males,highest level of relationships. Mean age of 56 indicates oldest people with highest level of education(2), highest income.hence
* Segment 0: Well -off
### For segment 1: Two-third are males, and almost all are singles,mean age 36,low education,lowest salary and resides in small cities.Hence
* Segment 1: Fewer opportunities
### Almost all in relationship with average age of 29, medium level of education, average income and average jobs.Evenly distributed amoung small,mid,and big cities.Hence 
* Segment 2: Career-focused
### Segment 3: Entirely men with few in relationships,low education with high income and occupation. Majority lives in big and mid cities. Hence
* Segment 3: Standard

### A cluster can be named in any way, but importantly it serves as reference.

```python
df_segm_analysis['N Obs'] = df_segm_kmeans[['Segment K-means','Sex']].groupby(['Segment K-means']).count()
```

```python
df_segm_analysis['Prop Obs'] = df_segm_analysis['N Obs']/df_segm_analysis['N Obs'].sum()
```

```python
df_segm_analysis
```

```python
df_segm_analysis.rename({0:'well-off',
                        1:'fewer-opportunities',
                        2:'standard',
                        3:'career focused'})
```

From the above, 13 percent of observation(the smallest) are in the well-off segment.Majority belong to the standard segment.

```python
df_segm_analysis.drop('N Obs', axis = 'columns',inplace=True)
```

```python
df_segm_analysis
```

```python
df_segm_kmeans['Labels'] = df_segm_kmeans['Segment K-means'].map({0:'well-off',
                                                                 1:'fewer opportunities',
                                                                 2:'standard',
                                                                 3:'career focused'})
```

```python
x_axis = df_segm_kmeans['Age']
y_axis = df_segm_kmeans['Income']
plt.figure(figsize = (10,8))
sns.scatterplot(x_axis, y_axis, hue = df_segm_kmeans['Labels'], palette = ['g','r','c','m'])
plt.show()
```

### From the plot the well-off segment is clearly distinguished from other segments. hence more processing.

## PCA

```python
pca = PCA()
```

```python
pca.fit(segmentation_std)
```

```python
pca.explained_variance_ratio_
```

### Select subset of components, while preserving variance with the help of graph. Rule of thumb: keep 80% of the vaiance.

```python
plt.figure(figsize = (12,9))
plt.plot(range(1,8), pca.explained_variance_ratio_.cumsum(), marker='o',linestyle='--')
plt.title('Explained Variance by Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulation')
```

```python
pca = PCA(n_components = 3)
```

```python
pca.fit(segmentation_std)
```

## PCA Results

```python
pca.components_
```

```python
df_pca_comp = pd.DataFrame(data = pca.components_,
                          columns = df_segmentation.columns.values,
                          index= ['Component 1','Component 2','Component 3'])
df_pca_comp
```

```python
sns.heatmap(df_pca_comp,
           vmin = -1,
           vmax=1,
           cmap = 'RdBu',
           annot = True)
plt.yticks([0,1,2],
          ['Component 1', 'Component 2','Component 3'],
          rotation = 45,
          fontsize = 9)
```

```python
pca.transform(segmentation_std)
```

```python
scores_pca = pca.transform(segmentation_std)
```

## K-means clustering with PCA

```python
wcss = []
for i in range(1,11):
    kmeans_pca = KMeans(n_clusters = i, init='k-means++', random_state= 42)
    kmeans_pca.fit(scores_pca)
    wcss.append(kmeans_pca.inertia_)
```

```python
plt.figure(figsize = (10,8))
plt.plot(range(1,11), wcss, marker = 'o', linestyle = '--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('K-means with Clustering')
plt.show()
```

```python
kmeans_pca = KMeans(n_clusters = 4, init='k-means++', random_state = 42)
```

```python
kmeans_pca.fit(scores_pca)
```

## K-means clustering with PCA Results

```python
df_segm_pca_kmeans = pd.concat([df_segmentation.reset_index(drop=True), pd.DataFrame(scores_pca)],axis =1)
df_segm_pca_kmeans.columns.values[-3: ] =['Component 1', 'Component 2','Component 3']
df_segm_pca_kmeans['Segment K-means PCA'] = kmeans_pca.labels_
```

```python
df_segm_pca_kmeans
```

```python
df_segm_pca_kmeans_freq = df_segm_pca_kmeans.groupby(['Segment K-means PCA']).mean()
df_segm_pca_kmeans_freq
```

* component 1: career
* component 2: Education lifestyle
* component 3: Experience

```python
df_segm_pca_kmeans_freq['N Obs'] = df_segm_pca_kmeans[['Segment K-means PCA','Sex']].groupby(['Segment K-means PCA']).count()
df_segm_pca_kmeans_freq['Prop Obs'] = df_segm_pca_kmeans_freq['N Obs'] / df_segm_pca_kmeans_freq['N Obs'].sum()
df_segm_pca_kmeans_freq = df_segm_pca_kmeans_freq.rename({0:'standard', 
                                                          1:'career focused',
                                                          2:'fewer opportunities', 
                                                          3:'well-off'})
df_segm_pca_kmeans_freq
```

```python
df_segm_pca_kmeans['Legend'] = df_segm_pca_kmeans['Segment K-means PCA'].map({0: 'standard',
                                                                             1: 'career focused',
                                                                             2: 'fewer opportunities',
                                                                             3:'well-off'})
```

```python
x_axis =df_segm_pca_kmeans['Component 2']
y_axis = df_segm_pca_kmeans['Component 1']
plt.figure(figsize =(10,8))
sns.scatterplot(x_axis,y_axis,hue = df_segm_pca_kmeans['Legend'],palette=['g','r','c','m'])
plt.title('Clusteres by PCA Components')
plt.show()
```

### We can now observer the separate clusters as compared to kmeans without PCA.Hence the biggest goal of PCA is to reduce features by combining them.
